```python
import pandas as pd
import numpy as np
import random
import vcf
from math import ceil
import pysam

def VCF_maker(path_to_VCF, chrom, start, end, K, N=None):
    reader = vcf.Reader(filename=path_to_VCF)
    reader.fetch(chrom,start=start,end=end)
    found_variants=[]
    for record in reader:
        found_variants.append(record)
    random.shuffle(found_variants)
    if N==None:
        N=int(len(found_variants)/K)
    count=0  
    if N > int(len(found_variants)/K):
        print(f'Найденное количество вариантов {len(found_variants)} не получится разбить по {N} штук в {K} vcf файлов')
        T=K
        while N > int(len(found_variants)/T):
            T-=1
        print(f'Поэтому будет уменьшено количество генерируемых vcf файлов c {K} до {T}')
        K=T
    for i in range(K):
        writer = vcf.Writer(open(f'output{i}.vcf', 'w'), reader)
        for variant in found_variants[i*N:(i+1)*N]:
            writer.write_record(variant)

        count+=1
        if count == K:
            break

VCF_maker('GSM4887638_D6.vcf.gz','chr1',10800000,10900000,5,5)
```
